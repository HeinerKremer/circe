```python
%load_ext autoreload
%autoreload 2
from plot_wandb import load_results
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
exp = 'dsprites_tricky'
methods = ['erm', 'circe', 'hscic', 'vmm', 'fgel']

results = load_results(exp, methods)
```

```python
data = []
for m in results.keys():
    for t in results[m].keys():
        for e in results[m][t]:
            data.append({'method': m, 'mode': t, 'res': e})
df = pd.DataFrame(data)
```

```python
_ = sns.barplot(data=df, x="method", y="res", hue="mode", estimator='median', errorbar="se")
_ = plt.title('dsprites tricky')
```
